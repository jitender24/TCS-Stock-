## Queery 2.9

```python
signals= pd.DataFrame(index=df.index)
signals['signal'] = 0.0
signals['21_SMA'] = df['Close Price'].rolling(window=21,min_periods=1).mean()
signals['34_SMA'] = df['Close Price'].rolling(window=34,min_periods=1).mean()
signals['signal'][21:] = np.where(signals['21_SMA'][21:] >signals['34_SMA'][21:],1.0,0.0)
signals['position']= signals['signal'].diff()
print(signals)
```

```python
fig = plt.figure(figsize=(30,20))
ax1 = fig.add_subplot(111, ylabel='Price in Rupees')

df['Close Price'].plot(ax=ax1,color='r',lw=2.)

signals[['21_SMA','34_SMA']].plot(ax=ax1,lw=2.)

ax1.plot(signals.loc[signals.position==1.0].index,
         signals['21_SMA'][signals.position==1.0],'^',
         markersize=20,color='g')

ax1.plot(signals.loc[signals.position==-1.0].index,
         signals['21_SMA'][signals.position==-1.0],'v',
         markersize=20,color='k')

plt.show()
```

## Query 2.10

```python
def bbands(price,length=14,numsd=2):
    ave = price.rolling(length).mean()
    std_dev = price.rolling(length).std()
    upper_band = ave + (std_dev*numsd)
    lower_band = ave - (std_dev*numsd)
    return np.round(ave,3),np.round(upper_band,3),np.round(lower_band,3)
df['ave'],df['upper'],df['lower'] = bbands(df['Close Price'])
```

```python
df['Average Price'].plot(c='k',figsize=(20,10),lw=3)
df['ave'].plot(c='k',figsize=(20,10),lw=1)
df['upper'].plot(c='g',figsize=(20,10),lw=1)
df['lower'].plot(c='r',figsize=(20,10),lw=1)
```
